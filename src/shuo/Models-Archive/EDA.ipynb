## Stock Filter
The only filters we applied is to only use tickers with at least 128 days of returns history at the time they are sampled, and removing tickers that had data issues with their adjusted closing prices as observed (tickers that had more than 10 days of over 100% or under -50% returns were removed, as these all had issues with their historical price data).

```python
# Filter data based on minimum days of history and data quality
min_days = 128
gain_threshold = 1.0
loss_threshold = -0.5
max_bad_days = 10

# Ensure each ticker has at least 128 days of returns
ticker_counts = filtered_data_3["TICKER"].value_counts()
valid_tickers = ticker_counts[ticker_counts >= min_days].index
filtered_data = filtered_data_3[filtered_data_3["TICKER"].isin(valid_tickers)]

# Filter stocks with more than 10 bad days
bad_days_count = filtered_data.groupby("TICKER")["RET"].apply(
    lambda x: ((x > gain_threshold) | (x < loss_threshold)).sum()
)
good_stocks = bad_days_count[bad_days_count <= max_bad_days].index
filtered_data = filtered_data[filtered_data["TICKER"].isin(good_stocks)]
```
