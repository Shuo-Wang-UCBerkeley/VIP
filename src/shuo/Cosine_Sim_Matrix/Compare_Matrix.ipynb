# Input Embedding Static Weights

```python
import numpy as np
import pandas as pd
import os

# Function to calculate Frobenius norm between two matrices
def frobenius_norm(matrix1, matrix2):
    return np.linalg.norm(matrix1 - matrix2)

# Load all cosine similarity matrices
matrix_files = ["cosine_sim_matrix_input_embedding_weights_100stocks_batchsize96_1feature11targets_1.csv", "cosine_sim_matrix_input_embedding_weights_100stocks_batchsize96_1feature11targets_2.csv", "cosine_sim_matrix_input_embedding_weights_100stocks_batchsize96_1feature11targets_3.csv"]  # Add all your files here
cosine_matrices = [pd.read_csv(file, index_col=0).values for file in matrix_files]

# Calculate Frobenius norm between successive matrices
frobenius_distances = [frobenius_norm(cosine_matrices[i], cosine_matrices[i+1]) for i in range(len(cosine_matrices)-1)]

# Plot the Frobenius distances
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(frobenius_distances) + 1), frobenius_distances, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Frobenius Distance')
plt.title('Convergence of Cosine Similarity Matrices')
plt.grid(True)
plt.show()

# Print Frobenius distances to see the values
for i, distance in enumerate(frobenius_distances, start=1):
    print(f"Distance between iteration {i} and {i+1}: {distance}")
```
